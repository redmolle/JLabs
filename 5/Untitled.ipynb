## Этапы поиска ip-адреса атаковавшей машины

### Шаг №1. Первичный поиск ip-адресов

Проведем поиск всех ip-адресов, содержащихся в логах:

```python
import re
path = 'sample.json'
pattern = r' [0-9]{1,3}[\.][0-9]{1,3}[\.][0-9]{1,3}[\.][0-9]{1,3}'
r = re.compile(pattern)

file = open(path)
json = file.read()
ip = r.findall(json)
ip = list(set(ip))
ip.sort()
print('--#  ip  #--')
for i in ip:
    print(i)
```

### Шаг №2. Аналитический разбор результатов

Мы сразу можем исключить такие адреса, как:

ip | причина
----------------- | ---
*0.0.0.0*         | *=*
*127.0.0.1*       | *localhost*
*192.168.1.1*     | *адрес локальной сети* (чаще всего шлюз)
*224.0.0.251*     | *маска подсети*
*224.0.0.252*     | *маска подсети*
*239.255.255.250* | *маска подсети*
*255.255.255.255* | *маска подсети*

А также все адреса, подходящий под следующий шаблон **172.18.39.XXX**, так как очевидно что это ip-адреса локальной сети.

Таким образом, методом исключения, мы достаточно быстро выяснили ip-адрес злоумышленника (**10.0.10.106**).
