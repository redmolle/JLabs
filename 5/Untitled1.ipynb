# Сбор и анализ логов скомпроментированной системы
Автор: Князев Константин  
Группа: БИСО-02-15

## Цель работы

Получить в ELK-стек логи из скомпроментированной системы и выявить факт несанкционированного проникновения.

## Системные требования

* ОС: Ubuntu 18.04.3 Server x64, Windows 7 x64
* Процессор: Intel Core i5
* RAM: 2+2Гб DDR3
* Более мощная конфигурация

## Этапы сбора и анализа логов системы

1. Сбор логов в ELK-стек
2. Получение логов и их отображение в Kibana
3. Анализ логов

### Шаг №1. Сбор логов в ELK-стек

Для передачи логов в стек ELK, можем дождаться автоматической передачи логов на сервер, но можно запустить сбор и отправку логов вручную, чтобы сэкономить время, для этого на источнике логов запустим *Командную строку* от имени администратора и выполним следующую команду

>> *"C:\Program Files\Winlogbeat\winlogbeat.exe"*

### Шаг №2. Первичный анали логов

После передачи логов из, предположительно, скомпроментированной системы, в logstash, мы можем просмотреть возможные логи, для этого перейдем в [Kibana][kibanaUrl] ![][pic1]
